```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = {
  'area': [1000, 1200, 1500, 2000, 2500, 3000],
  'price': [20000000,44000000, 90000000, 140000000, 550000000, 960000000 ]
}
df = pd.DataFrame(data)
print(df)
```

```python
%matplotlib inline
plt.xlabel('area(sq. feet)')
plt.ylabel('price(in rupees)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
area = [1225, 2425, 2625, 2885, 3165, 3355 ]
area2d = np.array([area]).reshape(-1,1)
p = pd.DataFrame(reg.predict(area2d))
print(p)
```

```python
%matplotlib inline
plt.label
plt.plot(area2d, p, color='blue')
plt.scatter(df.area, df.price, marker='+', color='red')
```
